```python
import dotenv
import os

dotenv.load_dotenv()
```

```python
key = os.environ["OPENAI_KEY"]
frames_dir = "data/frames/fold_dress"
```

```python
from utils.frameReader import FrameReader
from utils.predictor import GVLPredictor
from utils.visualiser import GVLVisualizer
from utils.vlm import ClaudeVLM, OpenAIVLM
```

```python
# Read frames
frames = FrameReader.read_frames_from_dir(frames_dir)

frames[0]
```

```python
# Initialize components
vlm = OpenAIVLM(key)
predictor = GVLPredictor(vlm)
visualizer = GVLVisualizer()
```

```python

# Get predictions
shuffled_preds, indices, unshuffled_preds = predictor.predict(
    frames=frames,
    task_description="Robot folding a black dress"
)
```

```python
len(shuffled_preds), len(indices), len(unshuffled_preds[1:]), len(frames)
```

```python
# Visualize
visualizer.plot_predictions(
    frames=frames[1:],
    shuffled_predictions=shuffled_preds,
    unshuffled_predictions=unshuffled_preds[1:],
    shuffle_indices=indices
)
```
